```python
import torchaudio
import os
import string
import torch
import torch.nn as nn
```

```python
folder = 'audio'
audio_extensions = ('.mp3', '.wav', '.flac')
```

```python
data = []
text_file = [
    "hello world",
    "how are you",
    "i am fine",
    "open chrome",
    "play music",
    "stop music",
    "what time is it",
    "close the window",
    "turn on wifi",
    "turn off wifi",
    "open youtube",
    "search google",
    "play some songs",
    "pause the video",
    "resume the video",
    "increase volume",
    "decrease volume",
    "open settings",
    "close application",
    "restart system"
]
```

```python
freq_mask = torchaudio.transforms.FrequencyMasking(15)
time_mask = torchaudio.transforms.TimeMasking(35)

for i, filename in enumerate(os.listdir(folder)):
    if filename.lower().endswith(audio_extensions):
        path = os.path.join(folder, filename)
        waveform, sampleRate = torchaudio.load(path)
        if len(waveform.shape) == 1:
            waveform = waveform.unsqueeze(0)
        if waveform.shape[0] > 1:
            waveform = waveform.mean(dim=0, keepdim=True)

        mel_spectrogram = torchaudio.transforms.MelSpectrogram(
            sample_rate=16000,
            n_mels=80,
        )(waveform)
        log_mel = torchaudio.transforms.AmplitudeToDB()(mel_spectrogram)
        log_mel = log_mel.squeeze(0)
        log_mel = log_mel.transpose(0, 1)
        log_mel = (log_mel - log_mel.mean()) / (log_mel.std() + 1e-5)

        log_mel = freq_mask(log_mel)
        log_mel = time_mask(log_mel)
        data.append((log_mel, text_file[i]))
```

```python
def preprocess_audio(path):
    waveform, sampleRate = torchaudio.load(path)
    
    # Standardize to Mono
    if waveform.shape[0] > 1:
        waveform = waveform.mean(dim=0, keepdim=True)
    
    # Resample if necessary (Model was trained on 16000Hz)
    if sampleRate != 16000:
        waveform = torchaudio.transforms.Resample(sampleRate, 16000)(waveform)
        
    # Generate Mel Spectrogram
    mel_spec = torchaudio.transforms.MelSpectrogram(sample_rate=16000, n_mels=80)(waveform)
    log_mel = torchaudio.transforms.AmplitudeToDB()(mel_spec)
    
    # Normalize (Crucial: Model expects this specific distribution)
    log_mel = log_mel.squeeze(0).transpose(0, 1)
    log_mel = (log_mel - log_mel.mean()) / (log_mel.std() + 1e-5)
    
    return log_mel.unsqueeze(0).to(device) # Shape: [1, Time, 80]
```

```python
chars = list(string.ascii_lowercase + " '")
char2idx = {c: i+1 for i, c in enumerate(chars)}
idx2char = {i: c for c, i in char2idx.items()}
```

```python
def encode(text):
    return [char2idx[c] for c in text.lower() if c in char2idx]
```

```python
class SimpleASR(nn.Module):
    def __init__(self, input_dim=80, hidden_dim=128, vocab_size=30):
        super().__init__()
        self.conv = nn.Conv1d(input_dim, hidden_dim, kernel_size=3, padding=1)

        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=4, batch_first=True),
            num_layers=2
        )

        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, x):
        x = x.transpose(1, 2)
        x = self.conv(x)
        x = x.transpose(1, 2)
        x = x.permute(1, 0, 2)
        x = self.transformer(x)
        x = self.fc(x)
        return x
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using Device: {device}")
```

```python
model = SimpleASR().to(device)
```

```python
from torch.nn.utils.rnn import pad_sequence

feature_list = []
targets_list = []
input_lengths = []
target_lengths = []

for features, text in data:
    feature_list.append(features)
    target = torch.tensor(encode(text))
    targets_list.append(target)

    input_lengths.append(features.size(0))
    target_lengths.append(len(target))

features_padded = pad_sequence(feature_list, batch_first=True)
targets_padded = pad_sequence(targets_list, batch_first=True)

features_padded = features_padded.to(device)
```

```python
output = model(features_padded)
ctc_loss = nn.CTCLoss(blank=0)
optimizer = torch.optim.Adam(model.parameters(), lr=5e-5)
```

```python
def decode(output):
    pred = torch.argmax(output, dim=-1).transpose(0, 1)
    pred = pred.squeeze(1).cpu().numpy()
    decoded_sentences = []

    for individual_pred in pred:
        text = ""
        prev = -1
        for p in individual_pred:
            if p != prev and p != 0:
                text += idx2char.get(p, "")
            prev = p
        decoded_sentences.append(text)
    return decoded_sentences
        
```

```python
# 1. Reset everything
feature_list = []
targets_list = []
target_lengths = []

# 2. Fill them up
for features, text in data:
    encoded = encode(text)
    if len(encoded) == 0:
        continue
    feature_list.append(features)
    target = torch.tensor(encode(text), dtype=torch.long)
    targets_list.append(target)
    target_lengths.append(len(target))

# 3. Create tensors
features_padded = pad_sequence(feature_list, batch_first=True).to(device)
targets_concat = torch.cat(targets_list).to(device)
target_lengths_tensor = torch.tensor(target_lengths, dtype=torch.long).to(device)

# 4. DEFINE batch_size based on the features we just made
current_batch_size = features_padded.size(0)

# 5. This will now be TRUE
print(f"Features Batch: {current_batch_size}")
print(f"Target Lengths: {len(target_lengths)}")
print(f"Match: {current_batch_size == len(target_lengths)}")

print("targets_concat shape:", targets_concat.shape)
print("target_lengths_tensor:", target_lengths_tensor)
print("sum target lengths:", target_lengths_tensor.sum().item())

batch_size = features_padded.size(0)

print("Batch size:", features_padded.size(0))
print("Target lengths size:", target_lengths_tensor.size(0))
print("Targets concat length:", targets_concat.size(0))
print("Sum target lengths:", target_lengths_tensor.sum().item())
```

```python
for i, features in enumerate(feature_list):
    print(f"Sample {i}: input_len={features.size(0)}, target_len={target_lengths[i]}")

for epoch in range(401):
    model.train()
    optimizer.zero_grad()

    output = model(features_padded)     
    output = output.permute(1, 0, 2)       

    time_steps = output.size(0)
    batch_size = output.size(1)
    input_lengths_tensor = torch.full(
        size=(batch_size,),
        fill_value=time_steps,
        dtype=torch.long
    ).to(device)

    loss = ctc_loss(
        output.log_softmax(2),
        targets_concat,
        input_lengths_tensor,
        target_lengths_tensor
    )

    loss.backward()
    optimizer.step()

    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

print("Training Complete!")
```
