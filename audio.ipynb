```python
from gtts import gTTS

phrases = [
    "hello world",
    "how are you",
    "i am fine",
    "open chrome",
    "play music",
    "stop music",
    "what time is it",
    "close the window",
    "turn on wifi",
    "turn off wifi",
    "open youtube",
    "search google",
    "play some songs",
    "pause the video",
    "resume the video",
    "increase volume",
    "decrease volume",
    "open settings",
    "close application",
    "restart system"
]

data = []

# 🔹 Generate SHORT audios (first 10)
for i, text in enumerate(phrases[:10]):
    tts = gTTS(text)
    filename = f"audio_short_{i}.mp3"
    tts.save(filename)
    
    data.append((filename, text))

# 🔹 Generate LONG audios (next 10)
for i in range(10):
    # combine 2 phrases randomly
    combined_text = phrases[i] + " " + phrases[i+10]
    
    tts = gTTS(combined_text)
    filename = f"audio_long_{i}.mp3"
    tts.save(filename)
    
    data.append((filename, combined_text))

print("Generated 20 audio files!")
```
